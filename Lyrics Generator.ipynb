```python
# f = open("time.txt","r+")
# data = f.read()
# f.close()
text_path = "time.txt"
def load_path(filename):
    with open(filename,encoding="utf8") as f:
        return f.read().lower()
```

```python
def getTransitionTable(data,k = 4):
    t = {}
    
    for i in range(len(data)-k):
        x = data[i:i+k]
        y = data[i+k]
        
        if t.get(x) is None:
            t[x]={}
            t[x][y] = 1
        else:
            if t[x].get(y) is None:
                t[x][y] = 1
            else:
                t[x][y] += 1
    return t
```

```python
text="Virat King"
T = getTransitionTable(data)
```

```python
def ConvertFreqIntoProb(t):
    for i in t.keys():
        s = float(sum(t[i].values()))
        for k in t[i].keys():
            t[i][k] = t[i][k]/s
            
    return t
```

```python
T = ConvertFreqIntoProb(T)
```

```python
def sample_next(sent,t,i,k=4):
    sentence = sent[-k:]
    if t.get(sentence) is None:
        return " "
    possible_chars = list(t[sentence].keys())
    possible_probabs = list(t[sentence].values())
    #np.random.seed(i)
    return np.random.choice(possible_chars, p = possible_probabs)
```

```python
def generateText(sent,t,k=4,max_length =2000):
    sentence = sent
    context = sentence[-k:]
    np.random.seed(11)
    for i in range(max_length):
        next_pred = sample_next(context,t,i,k)
        sentence += next_pred
        context = sentence[-k:]
    
    return sentence
```

```python
prediction = generateText("apna", T,max_length =2000)
```
